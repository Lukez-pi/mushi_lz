Investigating the $C$ matrix
==

```python
%matplotlib notebook
from mushi import SFS
import numpy as np
from matplotlib import pyplot as plt
from numpy.linalg import cond, svd
```

### The condition number $\kappa(A_n)$ for several values of $n$
Things go to ðŸ’© for $n$ larger than a couple hundred or so. This is a problem

```python
ns = np.logspace(1, 2.6, 100, dtype=np.int)
condition_numbers = [cond(SFS.C(n)) for n in ns]
plt.plot(ns, condition_numbers)
plt.xlabel('$n$')
plt.ylabel('$\kappa(C)$')
plt.xscale('log')
plt.yscale('log')
plt.show()
```

### Singular value spectra

```python
plt.figure()
for n in np.logspace(1, 2.6, 20, dtype=int):
    plt.plot(np.arange(1, n), svd(SFS.C(n))[1], label='$n = {}$'.format(n))
plt.ylabel('singular value')    
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.show()
```

### Just a colormap of the matrix

```python
n = 200
plt.matshow(SFS.C(n))
plt.colorbar()
plt.show()
```
